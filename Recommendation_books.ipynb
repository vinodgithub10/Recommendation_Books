# Model Building

## Popularity Based recommender system

```python
#segragating implicit and explict ratings datasets
explicit_rati = ratings_user_book[ratings_user_book['Book-Rating'] != 0]
implicit_rati = ratings_user_book[ratings_user_book['Book-Rating'] == 0]
```

```python
rating_count = pd.DataFrame(explicit_rati.groupby('Book-Title')['Book-Rating'].count())
rating_count.sort_values('Book-Rating', ascending=False).head()
```

```python
# Create column Rating average 
explicit_rati['Avg_Rating']=explicit_rati.groupby('Book-Title')['Book-Rating'].transform('mean')
# Create column Rating su
explicit_rati['Total_No_Of_Users_Rated']=explicit_rati.groupby('Book-Title')['Book-Rating'].transform('count')
```

```python
explicit_rati
```

### Weighted avearge rating formula(W) = (Rv+Cm)/(v+m)
- v is the number of votes for the books;
- m is the cutoff for the number of book ratings;
- R is the average rating of the book; and
- C is the mean vote across the whole report.
- Now we find the values of v,m,R,C.

```python
C=explicit_rati ['Avg_Rating'].mean()
m=explicit_rati['Total_No_Of_Users_Rated'].quantile(0.90)
popular_books = explicit_rati.loc[explicit_rati['Total_No_Of_Users_Rated'] >= m]
print(f'C={C} , m={m}')
print(popular_books.shape)
popular_books.head()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['Total_No_Of_Users_Rated']
    R = x['Avg_Rating']
    return ((R*v + C*m)/(v+m))
popular_books['Score'] = popular_books.apply(weighted_rating,axis=1)
 
 #Sorting books based on score calculated above
popular_books = popular_books.sort_values('Score', ascending=False)
popular_books
```

```python
#Keeping only one entry of each book
popular_books=popular_books.sort_values('Score', ascending=False).drop_duplicates('Book-Title').sort_index()
cm=sns.light_palette('yellow',as_cmap=True)
#Sorting books based on score calculated above
popular_books = popular_books.sort_values('Score', ascending=False)

#Printing the top 20 books
popular_books[['Book-Title', 'Total_No_Of_Users_Rated', 'Avg_Rating', 'Score']].reset_index(drop=True).head(20).style.background_gradient(cmap=cm)
```

#### Let's identify the books each user liked the most

```python
import numpy as np
rate = {}
rows_indexes = {}
for i,row in pt.iterrows():
    rows = [x for x in range(0,len(pt.T.columns))]
    combine = list(zip(row.index, row.values, rows))
    rated = [(x,y,z) for x,y,z in combine if (y==10)|(y==9)|(y==8)|(y==7)|(y==6)|(y==5)]
    index = sorted([(i[2],i[1]) for i in rated],key=lambda x: x[1],reverse=True)[:10]
    index1 = sorted([i[0] for i in index])
    row_names = [(i[0],i[1]) for i in rated]
    sort_ratings=sorted(row_names,key = lambda x:x[1],reverse=True)
    sort_rate = [i[0] for i in sort_ratings]
    rows_indexes[i] = index1[:4]
    rate[i] = sort_rate[:4]
```

```python
liked = {key:val for key,val in rate.items() if len(val)!=0 }
```

```python
liked_indexes = {key:value for key,value in rows_indexes.items() if len(value)!=0}
```

```python
from sklearn.neighbors import NearestNeighbors
```

### Unsupervised Nearest Neighbor Recommender

```python
n=5
#items_dic = {}
cosine_nn = NearestNeighbors(n_neighbors = n, algorithm='brute', metric = 'cosine')
item_cosine_nn = cosine_nn.fit(pt.T.values)
item_distances, item_indices = item_cosine_nn.kneighbors(pt.T.values)
```

### Item-Item based Recommender

```python
items_dic = {}
for i in range(len(pt.T.index)):
    item_idx = item_indices[i]
    col_names = pt.T.index[item_idx].tolist()
    items_dic[pt.T.index[i]] = col_names
```

### Recommendations

```python
toprecs = {}
for k,v in liked_indexes.items():
    item_idx = [j for i in item_indices[v] for j in i]
    item_dist = [j for i in item_distances[v] for j in i]
    combine = list(zip(item_dist, item_idx))
    diction = {i:d for d,i in combine if i not in v}
    zipped = list(zip(diction.keys(), diction.values()))
    sort = sorted(zipped, key = lambda x: x[1])
    recommendations = [(pt.columns[i],d) for i, d in sort]
    toprecs[k] = recommendations
```

```python
def recommendations(user, number_of_rec=5): 
    print("These are all the books you liked the most in the past: \n\n{}".format('\n'.join(liked[user])))
    print()
    print("We recommend to read these books too:\n")
    for k,v in toprecs.items():
        if user==k:
            for i in v[:number_of_rec]:
                print('{} with similarity: {:.4f}'.format(i[0], 1-i[1]))
```

```python
recommendations(11676)
```

## Evaluating the recommender system 

### Accuracy

```python
rate_all = {}
for i,row in df_final.iterrows():
    rows = [x for x in range(0,len(df_final.T.columns))]
    combine = list(zip(row.index, row.values, rows))
    rated = [(x,y,z) for x,y,z in combine if str(y)!='nan']
    row_names = [i[0] for i in rated]
    rate_all[i] = row_names
```
